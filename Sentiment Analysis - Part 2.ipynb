```python
from textblob import TextBlob

df['polarity'] = df.apply(lambda x: TextBlob(x['text']).sentiment.polarity, axis=1) #adds a new column with polarity scores
df['subjectivity'] = df.apply(lambda x: TextBlob(x['text']).sentiment.subjectivity, axis=1) #adds a new column with subjectivity scores

#citation - https://stackoverflow.com/questions/35559199/textblob-sentiment-analysis-on-a-csv-file
```

```python
df.head(5)
```

### 1) Average Subjectivity and Polarity Scores

```python
average_subjectivity_score = df['subjectivity'].mean() #calculates mean
average_polarity_score = df['polarity'].mean() #calculates mean

print(f' Average Subjectivity score is {average_subjectivity_score}')
print(f' Average Polarity score is {average_polarity_score}')
```

### 2) Visualizing Subjectivity and Polarity Scores on Histogram 

```python
#creates a histogram
import matplotlib.pyplot as plt

plt.hist(df['subjectivity'], bins=10) #, normed=1, alpha=0.75)

plt.xlabel('subjectivity score')
plt.ylabel('sentence count')
plt.grid(True)
plt.savefig('subjectivity.pdf')
plt.show()
```

```python
#creates a histogram
plt.hist(df['polarity'], bins=10) #, normed=1, alpha=0.75)

plt.xlabel('polarity score')
plt.ylabel('sentence count')
plt.grid(True)
plt.savefig('polarity.pdf')
plt.show()
```
